```python
import pytesseract
from PIL import Image
import gtts
```

```python
img = Image.open('/content/Screenshot 2024-02-29 233016.png')
print(img)
```

```python
result = pytesseract.image_to_string(img)
print(result)
tts = gtts.gTTS(result)
tts.save("hello.mp3")
```

```python
from IPython.display import Audio, display

# Replace '' with the name of your audio file
audio_file = 'hello.mp3'
display(Audio(audio_file, autoplay=True))
```
